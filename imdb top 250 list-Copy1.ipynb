```python
import requests
r = requests.get('https://www.imdb.com/chart/top')
print (r.text[0:1])
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
results = soup.find_all('td', attrs={'class':'titleColumn'})
ratings = soup.find_all('td', attrs={'class':'ratingColumn imdbRating'})
```

```python
records =[]
i = 0
for result in results:
    ranking = result.contents[0][7:-8]
    movie = result.find('a').text
    cast = result.find('a')['title']
    year = result.contents[3].text[1:-1]
    rating = ratings[i].text[1:-1]
    records.append((ranking, movie, cast, year, rating))
    i+=1
```

```python
records[0:3]
```

```python
import pandas as pd
df = pd.DataFrame(records, columns=['Rank', 'Movie', 'Cast', 'Year', 'IMDB Rating'])
```

```python
df.head()
```

```python
df.to_csv('imdb_250_movies.csv', index=False, encoding='utf-8')
```
